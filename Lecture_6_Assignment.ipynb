# E1
In the lithology data, add the description (description column) of all the formations in one string variable.

```
import pandas as pd
import nltk
from nltk import sent_tokenize
from nltk import word_tokenize
nltk.download('punkt')
litho_data=pd.read_csv('/content/strat_litho_overview.csv')
litho_data
```

```
text=''
for d in litho_data['lsuDescription']:
  text=text+d
```

```
len(text)
```

#E2 
Find the frequnecy distribution of the result obtained from E1

```
from nltk.probability import FreqDist
words=word_tokenize(text)
FreqDist(words)
```

# E3
Find the frequency distribution after removing the punctuation marks from the merged words

```
clean_word=[]
for w in words:
  if w.isalpha():
    clean_word.append(w)
FreqDist(clean_word)
```

# 
E4
Find the frequency distribution after removing the stop words from the result obtained from E3.

```
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words=stopwords.words('english')
new_clean_word=[]
for w in clean_word:
  if  w.lower() not in stop_words:
    new_clean_word.append(w.lower())
freq_words=FreqDist(new_clean_word)
freq_words.most_common(5)
```
